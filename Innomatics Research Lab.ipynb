```python
import pandas as pd
import sqlite3
import json
import os

print(f"Python is currently looking in: {os.getcwd()}")
print(f"Files in this folder: {os.listdir('.')}")

# Load orders
orders = pd.read_csv("orders.csv")

# Load users
with open("users.json", "r") as f:
    users_data = json.load(f)

users = pd.json_normalize(users_data)

# Create in-memory database
conn = sqlite3.connect(":memory:")

# Load SQL schema
with open(r"C:\Users\hp\Downloads\Project\restaurant.sql", "r") as f:
    sql_script = f.read()

conn.executescript(sql_script)

# Read restaurants table
restaurants = pd.read_sql_query("SELECT * FROM restaurants", conn)

# Merge datasets
merged_df = (
    orders
    .merge(users, on="user_id", how="left")
    .merge(restaurants, on="restaurant_id", how="left")
)

# Save final dataset
merged_df.to_csv("final_food_delivery_dataset.csv", index=False)

print("Final dataset created successfully!")

```
